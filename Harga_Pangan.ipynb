```python
#Using seaborn
var1 = 'Beras'
plt.figure(figsize=(4,3), dpi=100)
g = sns.histplot(df, x=var1, kde=True, color='steelblue')
g.set_title("Harga dan Jumlah Frekuensi Komoditas " + var1, fontsize=12)
g.set_xlabel("Harga " + var1, fontsize=12) #set x label
g.set_ylabel('Jumlah', fontsize=12, rotation=90) #set y label

data = (df['Beras'])

df3 = pd.DataFrame(data, columns=['Beras'])

# Menghitung jumlah observasi untuk setiap harga
jumlah_per_harga = df3['Beras'].value_counts()

# Mencari harga dengan jumlah yang paling banyak
harga_terbanyak = jumlah_per_harga.idxmax()
jumlah_terbanyak = jumlah_per_harga.max()

# Mencari harga dengan jumlah yang paling sedikit
harga_terdikit = jumlah_per_harga.idxmin()
jumlah_terdikit = jumlah_per_harga.min()

# Mencetak harga dengan jumlah yang paling banyak
print("Harga dengan jumlah terbanyak:", harga_terbanyak)
print("Jumlah terbanyak:", jumlah_terbanyak)

# Mencetak harga dengan jumlah yang paling sedikit
print("Harga dengan jumlah terdikit:", harga_terdikit)
print("Jumlah terdikit:", jumlah_terdikit)
```
